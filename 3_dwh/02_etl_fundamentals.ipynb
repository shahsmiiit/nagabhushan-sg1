
## What is ETL?

**ETL** stands for:
- **Extract**
- **Transform**
- **Load**

ETL is the **core backbone** of any data warehouse project.

Nearly **80–90% of data engineering effort** is spent designing and maintaining ETL pipelines.



### Extract

Extraction is the process of:
- Identifying required data from source systems
- Pulling data *as‑is* without modification

Common source systems:
- Databases
- CSV / Files
- APIs
- Event streams



### Transform

Transformation is the most complex ETL step.

It includes:
- Data cleansing
- Standardization
- Integration
- Business logic application
- Aggregations

This step converts raw data into **analysis‑ready data**.



### Load

Loading involves inserting transformed data into:
- Data warehouse tables
- Analytical layers

Load strategies:
- Full load
- Incremental load
- Append‑only
- Merge (Upsert)
