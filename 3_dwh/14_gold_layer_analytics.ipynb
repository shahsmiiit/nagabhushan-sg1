# Gold Layer Analytics — Assignment

## Context
The **gold layer** is analytics-ready and exposes a star schema via:

- `gold.dim_customers`
- `gold.dim_products`
- `gold.fact_sales`

Write **MySQL** queries to answer the questions below.

## Questions

1. **Retrieve the oldest customer (by age).**  
   Show: Show customer_key, customer_id, customer_number, first_name, last_name, birthdate, and age.

2. **Retrieve the youngest customer (by age).**  
   Show: Show customer_key, customer_id, customer_number, first_name, last_name, birthdate, and age..

3. **Compute revenue by category.**  
   Show: category and total revenue, sorted highest to lowest.

4. **Compute units sold by country.**  
   Show: country and total units sold, sorted highest to lowest.

5. **Find 3 customers with the fewest orders** (consider only customers with **≥ 1 order**).  
   Show: customer identifiers, name, and distinct order count.

6. **Find bottom 5 products by revenue.**  
   Show: product name and total revenue, sorted lowest to highest.

7. **Compute yearly product sales.**  
   Show: year, product name, and total revenue (product sales) for that year.
