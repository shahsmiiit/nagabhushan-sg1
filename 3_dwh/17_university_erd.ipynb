## University ERD Problem Statement



A university has multiple departments, each with a unique department code and department name.

Each department has multiple faculty members, each with a unique faculty ID, name, and email address.

Faculty members can take one course, and each course has a unique course code, course title, and credit value.

Each course is offered in a specific semester and each semester has a unique semester code, semester_startdate, semester_enddate.

Each semester has multiple students with unique student ID, name, contact number, email ID.


### Relationships:

A faculty member can take one course, and a course can be taught by one faculty

A department has multiple faculty members, and a faculty member belongs to one department

A course is offered in a specific semester, and a semester can offer multiple courses

A semester has multiple students, and a student can be enrolled in one semester

A course has multiple students, and a student can enroll in multiple courses
